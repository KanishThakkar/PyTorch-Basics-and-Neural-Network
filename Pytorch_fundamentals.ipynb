###Matrix Multiplication
1. Element wise multiplication
2. Matrix multiplication (Dot product)

###Matrix Multiplication Rules:
* Inner Dimention must match
e.x (2,3) @ (2,3) -> won't work
    (2,3) @ (3,2) -> work
* The resulten shape of matrix multiplication is its outer shape
e.x (2,3) @ (2,3) -> Output matrix shape = (2,2)

```python
#Element wise multiplication
print(TENSOR ,"*", TENSOR)
print("        ")
print(TENSOR * TENSOR)
```

To fix the tensor shape issue we can manipulate one of the tensor to be transpose.  

```python
tensor_a = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_b  = torch.tensor([[7,10],
                          [8,11],
                          [9,12]])
```

* torch.matmul(tensor_a,tensor_b) # give us error because it cant satisfied matrix mul. rules

```python
#TO fix this we use to transpose one of the matrix
print(tensor_b.T) # it will transpose the matrix from (3,2) -> (2,3)
print("")
print(tensor_b.T.shape)
print(" ")
print(torch.matmul(tensor_a,tensor_b.T))
```

### Reshaping , stacking , squezzing and unsquezzing Tensors
* Reshape - Reshape input tensor to defined shape
* View - Return the view of the input tensor but share same memory as input tensor
* stacking - combining multimple tensor on top of each other
* squezzing -remove all 1 dimention of tensor
* unsquezzing - add a 1 dimention of tensor
* permute - Return a view of tensor with dimention swapped in a certain way


```python
ex,ex.shape
```

```python
ex.reshape(1,6) # add an extra dimention..!
```

```python
z = ex.view(1,6) # here z is view of x
z, z.shape
```

```python
z[:,0] = 5  # here changes in z effect on x because z and x share same memory
z,ex
```

```python
# stack  - on top of each other
ex_stacked = torch.stack((ex,ex,ex,ex),dim = 0) # dim 0 means row wise stack
ex_stacked
```

```python
ex_stacked = torch.stack((ex,ex,ex,ex),dim = 1) # dim 1 means column wise stack
ex_stacked
```

```python
# squeeze method
print(f"privious arrray size :{ex.reshape(1,6).shape}")   # remove all 1 dimetnion from the original tensor shape
ex_squeezed = torch.squeeze(ex.reshape(1,6))
print("after squeezing")
print(f"New dimention are:{ex_squeezed.shape}")

#Unsqueeze method
print("after unsqueezing")
ex_unsqueezed = torch.unsqueeze(ex_squeezed,dim = 0)  # add 1 dimention by row or colum wise
print(f"unsqueeze dimention are{ex_unsqueezed.shape}")

# permute
x = torch.randn(3,4,6)
print(x.shape)
print(f"After permute :{x.permute(2,1,0).shape}")  #you can change tensor dimention by your self using permute
```

### Converting numpy array into pytorch tensor and converting back pytorch Tensor to numpy array
* Numpy array -> pytorch tensor   = torch.from_numpy(array)
* pytorch tensor -> numpy array = torch.numpy(array)


```python
import numpy as np
array = np.arange(1.0,6.0)
array_from_numpy = torch.from_numpy(array)
print(f"numpy array:{array}")
print(f"pytorch tensor:{array_from_numpy}")
print("")
print(f"dtype of numpy array:{array.dtype}")
print(f"dtype of pytorch tensor:{array_from_numpy.dtype}")
```

### Reproducibility (trying to take the random out of random)
How does this relate to neural networks and deep learning?

start with random numbers -> tensor operations -> try to make better (again and again and again)

For example, you create an algorithm capable of achieving X performance.
And then your friend tries it out to verify you're not crazy.
How could they do such a thing?
That's where reproducibility comes in.

* To reduce randomness in pytorch there is concept of `random seed`
Essentially what the random seed does is flavour the randomness

```python
A = torch.rand(2,3)
B = torch.rand(2,3)
print(A)
print(B)
print(A == B)   # so basically all time when you run it, it will give you diff diff output so reduce the randomness we use Random seef
```

```python
RANDOM_SEED = 42 #42 is ideal number for seeds , you can use whatever you want
torch.manual_seed(RANDOM_SEED)
C = torch.rand(2,3)
torch.manual_seed(RANDOM_SEED)
D = torch.rand(2,3)
print(C)
print(D)
print(C==D)
```

### Putting tensor on gpu


```python
# first set device name to cuda .cuda is gpu name for nvidia
device = "cuda"
tensor = torch.tensor([1,2,3])
# check if gpu is available
device = "cuda" if torch.cuda.is_available() else "cpu"
device

# count number of device on gpu
torch.cuda.device_count()

#Transfer this tensor from cpu to gpu
tensor_on_gpu = tensor.to(device)
tensor_on_gpu # we can convert tensor in numpy array when is in gpu.. to do this we need to convert back into cpu and then use it

# convert that tensor from gpu to cpu
tensor_on_cpu = tensor_on_gpu.cpu()
tensor_on_cpu.numpy()
```
