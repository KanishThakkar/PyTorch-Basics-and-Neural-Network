### Create multi layer Neural Network
Let’s create:

* Input layer: 2 neurons
* Hidden layer 1: 4 neurons
* Hidden layer 2: 3 neurons
* Output layer: 1 neuron
* With ReLU activation functions

```python
class Multilayer(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc1 = nn.Linear(2,4) # First hidden layer contains 2 input and has 4 neurons in 1st layer
    self.fc2 = nn.Linear(4,3) # Second hidden layer contains 3 neurons
    self.fc3 = nn.Linear(3,1) # Output layer which contain 1 neuron as an output
    self.relu = nn.ReLU  # Relu activation function

  def forward(self,x):
    x = self.relu(self.fc1(x));   # input → hidden layer 1 → ReLU
    x = self.relu(self.fc2(x));   # hidden layer 1 → hidden layer 2 → ReLU
    x = self.fc3(x);              # hidden layer 2 → output
    return x

# create a model
model  = Multilayer()
print(model)
```

```python
# Alternative and easy way for this
class multiLayerNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(2, 4),
            nn.ReLU(),
            nn.Linear(4, 3),
            nn.ReLU(),
            nn.Linear(3, 1)
        )

    def forward(self, x):
        return self.network(x)
model = multiLayerNN()
print(model)
```

### TO do backpropogation:
- step1: create a model
- step2: prepare the data
- step3: Define loss fun
- step4: choose optimizer
- step5:Make a training loop which contain all above data

```python
# step 1
model = multiLayerNN()

#prepare data
inputs = torch.tensor([[0.5, 1.0],
                       [1.0, 2.0],
                       [1.5, 2.5],
                       [2.0, 3.0]])

targets = torch.tensor([[1.0],
                        [2.0],
                        [3.0],
                        [4.0]])
# define loss fun
criterion = nn.MSELoss()

#Define optimizer
import torch.optim as optim
import torch.optim as optim

optimizer = optim.SGD(model.parameters(), lr=0.01)

# training loop with backpropogation
for epoch in range(100):

    # -------- FORWARD PASS --------
    outputs = model(inputs)         # predicted output
    loss = criterion(outputs, targets)   # compute loss

    # -------- BACKWARD PASS --------
    optimizer.zero_grad()   # reset previous gradients

    loss.backward()         # compute gradients

    optimizer.step()        # update weights

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

```
